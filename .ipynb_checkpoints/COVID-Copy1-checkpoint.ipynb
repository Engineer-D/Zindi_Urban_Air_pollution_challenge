   # STACKING

#### Try SDGRegressor,ridge regression,kernel ridge regression as level1 in STACKING SECTION

```python
from sklearn.linear_model import LinearRegression,SGDRegressor,RidgeCV,Ridge
from sklearn.ensemble import StackingRegressor
from sklearn.kernel_ridge import KernelRidge
level0=list()
level0.append(('xgb',XGBRegressor(max_depth=10,learning_rate=0.1,n_estimators=500,min_child_weight=10,objective='reg:linear',colsample_bytree=0.8,subsample=0.8,gamma=1)))
level0.append(('lgm',LGBMRegressor(boosting_type='dart',n_estimators=500,learning_rate=0.1,max_depth=10,num_leaves=1000,metric='mse')))
level0.append(('cat',CatBoostRegressor(n_estimators=1000,learning_rate=0.05,max_depth=10)))

level1= list()
level1.append(('lr',LinearRegression()))
level1.append(('kr',KernelRidge()))
level1.append(('SGD',SGDRegressor()))
level1.append(('rd',Ridge()))

model=StackingRegressor(estimators=level0,final_estimator=level1,cv=5)
model.fit(sX_train,y_train)
pred=model.predict(sX_test)
err=mean_squared_error(y_test,pred)**0.5
```
