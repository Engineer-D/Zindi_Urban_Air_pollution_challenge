# Base Model

```python
train_all = train.copy().dropna()
trains = train_all.loc[train_all.Date < '2020-04-01']
valid = train_all.loc[train_all.Date > '2020-04-01']
print(train.shape, valid.shape)
```

```python
in_cols = list(train.drop(['target','Place_ID X Date','Date',
                           'Place_ID','target_min','target_max',
                          'target_variance','target_count'], axis = 1))
target_cols = 'target'
```

```python
test_cols = list(test_gan_gan.columns)
train_cols = list(trains.columns)
```

```python
for i in in_cols:
    if i not in test_cols:
        print(i)
```

```python
test_gan_gan = test.drop(['Place_ID X Date','Date','Place_ID',
                         'L3_CH4_CH4_column_volume_mixing_ratio_dry_air',
                         'L3_CH4_aerosol_height','L3_CH4_aerosol_optical_depth',
                         'L3_CH4_sensor_azimuth_angle','L3_CH4_sensor_zenith_angle',
                         'L3_CH4_solar_azimuth_angle','L3_CH4_solar_zenith_angle',
                         'L3_SO2_solar_azimuth_angle','L3_SO2_solar_zenith_angle'], axis = 1)
```

```python
test.info()
```

```python
x_train, y_train = trains[in_cols], trains[target_cols]
x_valid, y_valid = valid[in_cols], valid[target_cols]
```

```python
lgb = LGBMRegressor()
```

```python
lgb.fit(x_train,y_train)
```

```python
pred = lgb.predict(x_valid)
```

```python
print(np.sqrt(mean_squared_error(y_valid,pred)))
```

```python
rf = RandomForestRegressor()
rf.fit(x_train,y_train)
rf_pred = rf.predict(x_valid)
print(np.sqrt(mean_squared_error(y_valid,rf_pred)))
```

```python
models = [CatBoostRegressor(),XGBRegressor(),LGBMRegressor(),
          RandomForestRegressor(),RidgeCV(),LinearRegression(),SVR()]
```

```python
stack = StackingCVRegressor(regressors = models,
                           meta_regressor= lgb, cv=12,
                            use_features_in_secondary=True,
                            store_train_meta_features=True,
                            shuffle=False)
```

```python
stack.fit(x_train, y_train)
```

```python
x_valided = x_valid.copy()
```

```python
x_valided.columns = ['f0', 'f1', 'f2', 'f3', 'f4', 'f5', 'f6', 'f7',
                  'f8', 'f9', 'f10', 'f11', 'f12','f13','f14','f15',
                 'f16','f17','f18','f19','f20','f21','f22', 'f23', 'f24',
                  'f25', 'f26', 'f27', 'f28', 'f29','f30', 'f31', 'f32', 
                  'f33', 'f34','f35','f36','f37','f38','f39','f40']
```

```python
stack_pred = stack.predict(x_valided)
```

```python
print(np.sqrt(mean_squared_error(y_valid,stack_pred)))
```
